1. Когда используется критерий Стьюдента, а когда Z –критерий?

Если неизвестна сигма генеральной совокупности и соблюдаются остальные условия применимости параметрических тестов, то следует использовать критерий Стьюдента.
Если сигма генеральной совокупности  известна и соблюдается нормальность распределения генеральной совокупности, то предпочтение отдают Z критерию.

2. Проведите тест гипотезы. Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм.

mu = 17 (средний диаметр шариков генеральной совокупности)
alpha = 0.05 = 5% (односторонний критерий)
n = 100 (шариков)
X = 17.5 (средний диаметр шариков выборки)
D(x) = 4 (дисперсия в кв.мм)
sigma = sqrt(D(x)) (среднее квадратичное отклонение, корень из дисперсии) = 2

Z=((X-μ)/σ/√n
Z = (17.5-17)/2 = 2.5

По таблице, для 0.95 Z = 1.695
Так как Z(фактическое) > Z(табличное) то принимаем альтернативную гипотезу.

4. Есть ли статистически значимые различия в росте дочерей?

Рост матерей 172, 177, 158, 170, 178,175, 164, 160, 169, 165

Рост взрослых дочерей: 173, 175, 162, 174, 175, 168, 155, 170, 160

```python
import numpy as np 
import scipy.stats as stats

x = np.array([172, 177, 158, 170, 178, 175, 164, 160, 169, 165])
y = np.array([173, 175, 162, 174, 175, 168, 155, 170, 160])
stats.ttest_ind(x,y)
```

Т.к. значение pvalue=0.810, слишком высокое, значение a=0.005, мы можем сделать вывод, что статистически значимых различий в росте дочерей нет.
